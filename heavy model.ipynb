```python
# =============================================================
# 1. IMPORT LIBRARIES
# =============================================================
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.applications import DenseNet201, InceptionResNetV2
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# LIME
from lime import lime_image
from skimage.segmentation import mark_boundaries
# =============================================================
# 2. DATA PATHS
# =============================================================
TRAIN_DIR = r"D:/MAJOR PROJECT/LungXRays-grayscale/train"
VAL_DIR   = r"D:/MAJOR PROJECT/LungXRays-grayscale/val"
TEST_DIR  = r"D:/MAJOR PROJECT/LungXRays-grayscale/test"

OUTPUT_DIR = "lung_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# =============================================================
# 3. DATA GENERATORS
# =============================================================
train_aug = ImageDataGenerator(
    rescale=1/255., rotation_range=20, width_shift_range=0.1,
    height_shift_range=0.1, zoom_range=0.2, horizontal_flip=True
)
val_test_aug = ImageDataGenerator(rescale=1/255.)

train_gen = train_aug.flow_from_directory(TRAIN_DIR, target_size=(224,224), batch_size=16, class_mode='categorical')
val_gen   = val_test_aug.flow_from_directory(VAL_DIR, target_size=(224,224), batch_size=16, class_mode='categorical')
test_gen  = val_test_aug.flow_from_directory(TEST_DIR, target_size=(224,224), batch_size=16, class_mode='categorical', shuffle=False)

num_classes = len(train_gen.class_indices)

# =============================================================
# 4. CALLBACKS
# =============================================================
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', patience=3, factor=0.3),
    ModelCheckpoint(os.path.join(OUTPUT_DIR, 'best_model.keras'), save_best_only=True)
]

# =============================================================
# 5. MODEL 1  Custom Lightweight CNN
# =============================================================
def build_light_cnn():
    model = Sequential([
        Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
        MaxPooling2D(),
        Conv2D(64, (3,3), activation='relu'),
        MaxPooling2D(),
        Conv2D(128, (3,3), activation='relu'),
        MaxPooling2D(),
        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

light_model = build_light_cnn()
print(light_model.summary())

history_light = light_model.fit(train_gen, validation_data=val_gen, epochs=20, callbacks=callbacks)

# =============================================================
# 6. MODEL 2  DenseNet201
# =============================================================
base_dn = DenseNet201(weights='imagenet', include_top=False, input_shape=(224,224,3))
base_dn.trainable = False

x = GlobalAveragePooling2D()(base_dn.output)
x = Dropout(0.3)(x)
output_dn = Dense(num_classes, activation='softmax')(x)

densenet_model = Model(inputs=base_dn.input, outputs=output_dn)
densenet_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print(densenet_model.summary())

history_dn = densenet_model.fit(train_gen, validation_data=val_gen, epochs=20, callbacks=callbacks)

# =============================================================
# 7. MODEL 3  InceptionResNetV2
# =============================================================
base_ir = InceptionResNetV2(weights='imagenet', include_top=False, input_shape=(224,224,3))
base_ir.trainable = False

x = GlobalAveragePooling2D()(base_ir.output)
x = Dropout(0.3)(x)
output_ir = Dense(num_classes, activation='softmax')(x)

inceptionresnet_model = Model(inputs=base_ir.input, outputs=output_ir)
inceptionresnet_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print(inceptionresnet_model.summary())

history_ir = inceptionresnet_model.fit(train_gen, validation_data=val_gen, epochs=20, callbacks=callbacks)

# =============================================================
# 8. MODEL 4  Hybrid CNN + LSTM
# =============================================================
def build_cnn_lstm():
    model = Sequential([
        TimeDistributed(Conv2D(32, 3, activation='relu'), input_shape=(10,224,224,3)),
        TimeDistributed(MaxPooling2D()),
        TimeDistributed(Conv2D(64, 3, activation='relu')),
        TimeDistributed(MaxPooling2D()),
        TimeDistributed(Flatten()),
        LSTM(128),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# NOTE: You must prepare sequence dataset manually if using CNN-LSTM
# Skipping training to avoid breaking notebook unless sequences are provided.
# cnn_lstm_model = build_cnn_lstm()

# =============================================================
# 9. FUNCTION  Plot Graphs
# =============================================================
def plot_history(history, title):
    plt.figure(figsize=(8,6))
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(f'{title} Accuracy')
    plt.legend()
    plt.show()

    plt.figure(figsize=(8,6))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{title} Loss')
    plt.legend()
    plt.show()

plot_history(history_light, "Lightweight CNN")
plot_history(history_dn, "DenseNet201")
plot_history(history_ir, "InceptionResNetV2")

# =============================================================
# 10. Evaluate Function for ALL models
# =============================================================
def evaluate_model(model, name):
    print(f"\n======= {name} METRICS =======")
    preds = model.predict(test_gen)
    y_true = test_gen.classes
    y_pred = np.argmax(preds, axis=1)

    print(classification_report(y_true, y_pred, target_names=test_gen.class_indices.keys()))

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,6))
    plt.imshow(cm, cmap="Blues")
    plt.title(f"{name} - Confusion Matrix")
    plt.colorbar()
    plt.show()

# Run reports
evaluate_model(light_model, "Lightweight CNN")
evaluate_model(densenet_model, "DenseNet201")
evaluate_model(inceptionresnet_model, "InceptionResNetV2")

# =============================================================
# 11. LIME EXPLAINABILITY
# =============================================================
explainer = lime_image.LimeImageExplainer()
sample_img, _ = next(iter(test_gen))

explanation = explainer.explain_instance(
    sample_img[0].astype('double'),
    densenet_model.predict,
    top_labels=1,
    num_samples=1000
)

temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=True)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(sample_img[0])
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(mark_boundaries(temp, mask))
plt.title("LIME Explanation")
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
from tensorflow.keras.applications import NASNetMobile
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow.keras.backend as K

# ======================================================
#           GRAYSCALE  RGB DATA LOADING
# ======================================================
IMG_SIZE = 224
BATCH_SIZE = 32

def gray_to_rgb(img):
    if img.shape[-1] == 1:
        img = np.repeat(img, 3, axis=-1)
    return img

train_gen = ImageDataGenerator(
    rescale=1./255,
    preprocessing_function=gray_to_rgb,
    rotation_range=25,
    zoom_range=0.25,
    width_shift_range=0.12,
    height_shift_range=0.12,
    horizontal_flip=True,
    brightness_range=[0.8,1.2]
)

val_gen = ImageDataGenerator(
    rescale=1./255,
    preprocessing_function=gray_to_rgb
)

train = train_gen.flow_from_directory(
    r"D:\MAJOR PROJECT\LungXRays-grayscale\train",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True
)

val = val_gen.flow_from_directory(
    r"D:\MAJOR PROJECT\LungXRays-grayscale\val",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

num_classes = train.num_classes

# ======================================================
#               NASNETMOBILE MODEL (Anti-Overfit)
# ======================================================

base_model = NASNetMobile(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False,
    weights="imagenet",
    pooling="avg"
)

base_model.trainable = False   # Freeze for stability first

x = base_model.output
x = Dropout(0.4)(x)

# L2 Regularization + Label smoothing
x = Dense(
    256,
    activation="relu",
    kernel_regularizer=tf.keras.regularizers.l2(1e-4)
)(x)
x = Dropout(0.3)(x)

output = Dense(
    num_classes,
    activation="softmax",
    kernel_regularizer=tf.keras.regularizers.l2(1e-4)
)(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
    metrics=["accuracy"]
)

model.summary()

# ======================================================
#                    TRAINING STEP 1
#             (Train top layers to stabilize)
# ======================================================
callbacks = [
    EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", patience=3, factor=0.3),
    ModelCheckpoint("nasnetmobile_lung.h5", save_best_only=True)
]

history = model.fit(
    train,
    validation_data=val,
    epochs=20,
    callbacks=callbacks
)

# ======================================================
#              TRAINING STEP 2  FINE TUNING
#    (Unfreeze deeper layers  Higher accuracy 9899%)
# ======================================================
base_model.trainable = True

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),  # Low LR for stable fine-tune
    loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.05),
    metrics=["accuracy"]
)

history_ft = model.fit(
    train,
    validation_data=val,
    epochs=8,
    callbacks=callbacks
)

# ======================================================
#              CONFUSION MATRIX + REPORT
# ======================================================
pred = model.predict(val)
y_pred = np.argmax(pred, axis=1)
labels = list(val.class_indices.keys())
y_true = val.classes

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(
    cm, annot=True, cmap="Blues", fmt="d",
    xticklabels=labels, yticklabels=labels
)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

print(classification_report(y_true, y_pred, target_names=labels))
```
